```python
df = pd.read_csv('Heart_Disease_Prediction.csv')
#df
```

```python
df_men = df[df['Sex'] == 1].reset_index(drop=True)
#df_men
#len(df_men)
```

```python
df_men_pres = df_men[df_men['Heart Disease'] == 'Presence'].reset_index(drop = True)
#df_men_pres
#len(df_men_pres)
```

```python
class Cholesterol_Line(object):
    def __init__(self, Chol_Line):
        self.Chol_Line = Chol_Line
        self.above_dict = {}
        self.below_dict = {}
    def above_below(self, x):
        for i in range(0, len(x)):
            if x['Cholesterol'][i] > self.Chol_Line:
                self.above_dict.update({x['Age'][i] : x['Cholesterol'][i]})

            else:
                self.below_dict.update({x['Age'][i] : x['Cholesterol'][i]})
    
    def main(self, x):
        global df_above
        global df_below
        self.above_below(x)
        df_above = pd.DataFrame(list(self.above_dict.items()),
                                    columns = ['Age', 'Cholesterol'])
        df_below = pd.DataFrame(list(self.below_dict.items()),
                                     columns = ['Age', 'Cholesterol'])
        
```

```python
Chol_men = Cholesterol_Line(200)
```

```python
Chol_men.main(df_men)
```

```python
df_men_below = df_below
#df_men_below
```
