```python
class Cholesterol_Line(object):
    def __init__(self, df, Chol_Line: float, gender: str, column_1: str, column_2: str, Pres_or_Abs: str):
        self.df = df.reset_index(drop = True)
        self.Chol_Line = Chol_Line
        self.gender = gender
        self.column_1 = column_1
        self.column_2 = column_2
        self.Pres_or_Abs = Pres_or_Abs

    def gender_filter(self,df, gender):
        global df_gender
        if gender == 'man':
            self.df_gender = df[df['Sex'] == 1].reset_index(drop=True)
        elif gender == 'woman':
            self.df_gender = df[df['Sex'] == 0].reset_index(drop=True)
        else:
            
            print("Error! Use man or woman!")
            
    def heart_disease_gender(self, df_gender):
        global df_presence
        global df_absence
        if self.Pres_or_Abs == 'Presence':
            df_presence = self.df_gender[self.df_gender['Heart Disease'] == self.Pres_or_Abs].reset_index(drop = True)
            df_absence = self.df_gender[self.df_gender['Heart Disease'] != self.Pres_or_Abs].reset_index(drop = True)
        elif self.Pres_or_Abs == 'Absence':
            df_presence = self.df_gender[self.df_gender['Heart Disease'] != self.Pres_or_Abs].reset_index(drop = True)
            df_absence = self.df_gender[self.df_gender['Heart Disease'] == self.Pres_or_Abs].reset_index(drop = True)
        else:
            print("Error! Use Presence or Absence")
        
    def chol_line(self, Chol_Line, column_1, column_2):

        global df_chol_line_above
        global df_chol_line_below
        global df_gender_above
        global df_gender_below
        global df_presence_above
        global df_presence_below
        
        df_chol_line_above = df[df[self.column_2] >= self.Chol_Line].reset_index(drop = True)
        df_chol_line_below = df[df[self.column_2] < self.Chol_Line].reset_index(drop = True)
        df_gender_above = self.df_gender[self.df_gender[self.column_2] >= self.Chol_Line].reset_index(drop = True)
        df_gender_below = self.df_gender[self.df_gender[self.column_2] < self.Chol_Line].reset_index(drop = True)
        df_presence_above = df_presence[df_presence[self.column_2] >= self.Chol_Line].reset_index(drop = True)
        df_presence_below = df_presence[df_presence[self.column_2] < self.Chol_Line].reset_index(drop = True)

        

    def main(self):
        self.gender_filter(self.df, self.gender)
        self.heart_disease_gender(self.df_gender)
        self.chol_line(self.Chol_Line, self.column_1, self.column_2)
```

```python
class Bayes(object):
    def __init__(self):
        self.Pres_Above = 0
        self.Above = 0
        self.Pres_Below = 0
        self.Below = 0
        self.Pres = 0
        self.Pres_Above_len = 0
        self.Above_len = 0
        self.Pres_Below_len = 0
        self.Below_len = 0
        self.Pres_len = 0
        self.One_Percent = 0
        
    def df_length(self, a, b, c, d, e):
        self.Pres_Above_len = len(a)
        self.Above_len = len(b)
        self.Pres_Below_len = len(c)
        self.Below_len = len(d)
        self.Pres_len = len(e)
        self.One_Percent = len(e)/100

        #percentage calculation 
        self.Pres_Above = len(a)/self.One_Percent
        self.Above = len(b)/self.One_Percent
        self.Pres_Below = len(c)/self.One_Percent
        self.Below = len(d)/self.One_Percent

    def Bayes_calculate(self, a, b, c, d, e):
        global Bayes_calculated
        self.df_length(a, b, c, d, e)
        Bayes_calculated = ((self.Pres_Above*self.Above) / ((self.Pres_Above*self.Above) + (self.Pres_Below*self.Below)))*100
        #return Bayes_calculated
```
