```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Heart_Disease_Prediction.csv')
#df
```

```python
class Cholesterol_Line(object):
    def __init__(self, df, Chol_Line: float, gender: str, column_1: str, column_2: str, Pres_or_Abs: str):
        self.df = df.reset_index(drop = True)
        self.Chol_Line = Chol_Line
        self.gender = gender
        self.column_1 = column_1
        self.column_2 = column_2
        self.Pres_or_Abs = Pres_or_Abs
        self.df_gender = []
        self.df_presence = []
        self.df_absence = []
        self.df_gender_presence = []
        self.df_gender_absence = []
        self.df_above = []
        self.df_below = []
        self.df_gender_above = []
        self.df_gender_below = []
        self.df_presence_above = []
        self.df_presence_below = []
        self.df_absence_above = []
        self.df_absence_below = []
        self.df_gender_presence_above = []
        self.df_gender_presence_below = []
        self.df_gender_absence_above = []
        self.df_gender_absence_below = []

    def df_main(self, df):
        global df_global
        df_global = self.df

    def heart_disease_presence_df_main(self, df):
        global df_presence_global
        global df_absence_global
        
        if self.Pres_or_Abs == 'Presence':
            self.df_presence = self.df[self.df['Heart Disease'] == self.Pres_or_Abs].reset_index(drop = True)
            self.df_absence = self.df[self.df['Heart Disease'] != self.Pres_or_Abs].reset_index(drop = True)
        elif self.Pres_or_Abs == 'Absence':
            self.df_presence = self.df[self.df['Heart Disease'] != self.Pres_or_Abs].reset_index(drop = True)
            self.df_absence = self.df[self.df['Heart Disease'] == self.Pres_or_Abs].reset_index(drop = True)
        else:
            print("Error! Use Presence or Absence")

        df_presence_global = self.df_presence
        df_absence_global = self.df_absence

    def gender_filter_df_main(self,df, gender):
        global df_gender_global
        
        if gender == 'man':
            self.df_gender = self.df[self.df['Sex'] == 1].reset_index(drop=True)
        elif gender == 'woman':
            self.df_gender = self.df[self.df['Sex'] == 0].reset_index(drop=True)
        else:
            print("Error! Use man or woman!")

        df_gender_global = self.df_gender
        
    def heart_disease_gender(self, df_gender):
        global df_gender_presence_global
        global df_gender_absence_global
        
        if self.Pres_or_Abs == 'Presence':
            self.df_gender_presence = self.df_gender[self.df_gender['Heart Disease'] == self.Pres_or_Abs].reset_index(drop = True)
            self.df_gender_absence = self.df_gender[self.df_gender['Heart Disease'] != self.Pres_or_Abs].reset_index(drop = True)
        elif self.Pres_or_Abs == 'Absence':
            self.df_gender_presence = self.df_gender[self.df_gender['Heart Disease'] != self.Pres_or_Abs].reset_index(drop = True)
            self.df_gender_absence = self.df_gender[self.df_gender['Heart Disease'] == self.Pres_or_Abs].reset_index(drop = True)
        else:
            print("Error! Use Presence or Absence")

        df_gender_presence_global = self.df_gender_presence
        df_gender_absence_global = self.df_gender_absence
        
    def chol_line(self, Chol_Line, column_1, column_2):

        global df_above_global
        global df_below_global
        global df_gender_above_global
        global df_gender_below_global
        global df_presence_above_global
        global df_presence_below_global
        global df_absence_above_global
        global df_absence_below_global
        global df_gender_presence_above_global
        global df_gender_presence_below_global
        global df_gender_absence_above_global
        global df_gender_absence_below_global
        
        self.df_above = self.df[self.df[self.column_2] >= self.Chol_Line].reset_index(drop = True)
        df_above_global = self.df_above
        self.df_below = self.df[self.df[self.column_2] < self.Chol_Line].reset_index(drop = True)
        df_below_global = self.df_below
        
        self.df_gender_above = self.df_gender[self.df_gender[self.column_2] >= self.Chol_Line].reset_index(drop = True)
        df_gender_above_global = self.df_gender_above
        self.df_gender_below = self.df_gender[self.df_gender[self.column_2] < self.Chol_Line].reset_index(drop = True)
        df_gender_below_global = self.df_gender_below
        
        self.df_presence_above = self.df_presence[self.df_presence[self.column_2] >= self.Chol_Line].reset_index(drop = True)
        df_presence_above_global = self.df_presence_above
        self.df_presence_below = self.df_presence[self.df_presence[self.column_2] < self.Chol_Line].reset_index(drop = True)
        df_presence_below_global = self.df_presence_below
        self.df_absence_above = self.df_absence[self.df_absence[self.column_2] >= self.Chol_Line].reset_index(drop = True)
        df_absence_above_global = self.df_absence_above
        self.df_absence_below = self.df_absence[self.df_absence[self.column_2] < self.Chol_Line].reset_index(drop = True)
        df_absence_below_global = self.df_absence_below
        
        self.df_gender_presence_above = self.df_gender_presence[self.df_gender_presence[self.column_2] >= 
            self.Chol_Line].reset_index(drop = True)
        df_gender_presence_above_global = self.df_gender_presence_above
        self.df_gender_presence_below = self.df_gender_presence[self.df_gender_presence[self.column_2] < 
            self.Chol_Line].reset_index(drop = True)
        df_gender_presence_below_global = self.df_gender_presence_below
        self.df_gender_absence_above = self.df_gender_absence[self.df_gender_absence[self.column_2] >= 
            self.Chol_Line].reset_index(drop = True)
        df_gender_absence_above_global = self.df_gender_absence_above
        self.df_gender_absence_below = self.df_gender_absence[self.df_gender_absence[self.column_2] < 
            self.Chol_Line].reset_index(drop = True)
        df_gender_absence_below_global = self.df_gender_absence_below

        


    def print_out(self):
        print("Available dataframe(name) and lenght: " + "\n" + "gender: " + self.gender + 
              "\n" + "df: " + str(len(self.df)) +
              "\n" + "df_presence: " + str(len(self.df_presence)) +
              "\n" + "df_absence: " + str(len(self.df_absence)) +
              "\n" + "df_gender: " + str(len(self.df_gender)) + 
              "\n" + "df_gender_presence: " + str(len(self.df_gender_presence)) +
              "\n" + "df_gender_absence: " + str(len(self.df_gender_absence)) + 
              "\n" + "df_above: " + str(len(self.df_above)) +
              "\n" + "df_below: " + str(len(self.df_below)) + 
              "\n" + "df_gender_above: " + str(len(self.df_gender_above)) +
              "\n" + "df_gender_below: " + str(len(self.df_gender_below)) + 
              "\n" + "df_presence_above: " + str(len(self.df_presence_above)) +
              "\n" + "df_presence_below: " + str(len(self.df_presence_below)) + 
              "\n" + "df_gender_presence_above: " + str(len(self.df_gender_presence_above)) +
              "\n" + "df_gender_presence_below: " + str(len(self.df_gender_presence_below)) + 
              "\n" + "df_gender_absence_above: " + str(len(self.df_gender_absence_above)) + 
              "\n" + "df_gender_absence_below: " + str(len(self.df_gender_absence_below)) + 
              "\n" + "Add _global at the end of the name of the dataframe to acces it! For Example: df_global." 
             )
        

    def main(self):
        self.df_main(self.df)
        self.heart_disease_presence_df_main(self.df)
        self.gender_filter_df_main(self.df, self.gender)
        self.heart_disease_gender(self.df_gender)
        self.chol_line(self.Chol_Line, self.column_1, self.column_2)
        self.print_out()
```

```python
chol_men = Cholesterol_Line(df, 200, 'man', 'Age', 'Cholesterol', 'Presence')
```

```python
chol_men.main()
```

```python
class Confusion_Matrix(object):

    def __init__(self, a, b, c, d, shape, title : str):
        self.a = a
        self.b = b
        self.c = c
        self.d = d
        self.shape = shape
        self.title = title
        self.confusion_list = []
        self.confusion_numpy_array = []
        self.confusion_matrix = []
        self.lenght_a = 0
        self.length_b = 0
        self.length_c = 0
        self.length_d = 0

    def length_calculation(self):

        self.length_a = len(self.a)
        self.length_b = len(self.b)
        self.length_c = len(self.c)
        self.length_d = len(self.d)
        self.confusion_list = [self.length_a, self.length_b, self.length_c, self.length_d]

    def confusion_matrix_calculation(self):
        global confusion_matrix_global
        
        self.confusion_numpy_array = np.array(self.confusion_list)
        self.confusion_matrix = self.confusion_numpy_array.reshape(self.shape)
        confusion_matrix_global = self.confusion_matrix
        
    def print_out(self):
        print('The confusion matrix of the ' + self.title + ': ' + '\n')
        print(self.confusion_matrix)

    def main(self):
        self.length_calculation()
        self.confusion_matrix_calculation()
        self.print_out()

```

```python
confusion_matrix_men = Confusion_Matrix(df_gender_presence_above_global, df_gender_absence_above_global, 
                                        df_gender_presence_below_global, df_gender_absence_below_global,
                                        (2, 2), "Cholesterol and Heart Disease")
```

```python
confusion_matrix_men.main()
```
