```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Heart_Disease_Prediction.csv')
#df
```

```python
df_men = df[df['Sex'] == 1].reset_index(drop=True)
#df_men
len(df_men)
```

```python
df_men_pres = df_men[df_men['Heart Disease'] == 'Presence'].reset_index(drop = True)
#df_men_pres
len(df_men_pres)
```

```python
df_men_chol = df_men[['Age', 'Cholesterol']]
df_men_dict = df_men_chol.set_index('Age').to_dict('list')['Cholesterol']
df_men_dict
#df_men_chol
```

```python
len(df_men_dict)
```

```python
df_men_dict = dict(zip(df_men_chol.Age, df_men_chol.Cholesterol))
df_men_dict
```

```python
len(df_men_dict)
```

```python
df_men_chol_list = df_men_chol.values.tolist()
len(df_men_chol_list)
```

```python
d = {}
for k, v in df_men_chol_list:
    d.setdefault(k, []).append(v)
sorted(d.items())
```

```python
df_chol = pd.DataFrame(list(d.items()), columns=['Age', 'Cholesterol'])
df_chol
```

```python
df_chol = pd.DataFrame.from_dict(d, orient='index')
df_chol
```

```python
df_chol = pd.DataFrame.from_dict(d, orient='index').T.melt(var_name='Age', 
                                                           value_name='Cholesterol').dropna(subset=['Cholesterol']).reset_index(drop=True)
df_chol
```
